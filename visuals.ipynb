```python
import numpy as np
import pandas as pd

dataset = r'C:\Users\Professor Moruf\Downloads\telecom_churn.csv'

# set the decimal no to 2dp
pd.set_option("display.precision", 2)

df = pd.read_csv(dataset)


import matplotlib.pyplot as plt
import seaborn as sns

# Set the style of seaborn
sns.set(style="whitegrid")

# Visualization 1: Distribution of Churn
plt.figure(figsize=(8, 6))
sns.countplot(x='Churn', data=df, palette='Set1')
plt.title('Distribution of Churn')
plt.xlabel('Churn')
plt.ylabel('Count')
plt.show()

```

The initial analysis provides a high-level view of churn in the dataset. Understanding the churn rate and exploring contributing factors are crucial for making informed business decisions.

```python


# Visualization 2: Average Day Minutes for Churned and Loyal Users
plt.figure(figsize=(10, 6))
sns.barplot(x='Churn', y='Total day minutes', data=df, palette='pastel')
plt.title('Average Day Minutes for Churned and Loyal Users')
plt.xlabel('Churn')
plt.ylabel('Average Day Minutes')
plt.show()


```


Conclusion:

There is a noticeable difference in the average day minutes between churned and loyal users.
The higher average day minutes for churned users might indicate that they were more active or engaged with the telecom services during the day.
This could be an important feature contributing to churn, as high usage might be associated with dissatisfaction or specific service issues.

```python
# Visualization 3: Correlation Heatmap
plt.figure(figsize=(12, 10))

# Exclude non-numeric columns from correlation calculation
numeric_df = df.select_dtypes(include=['number'])
correlation_matrix = numeric_df.corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()

```

Conclusion:

The heatmap provides insights into the relationships between different numeric features in the dataset.
Positive correlations (closer to 1) suggest that as one variable increases, the other tends to increase as well. Negative correlations (closer to -1) suggest that as one variable increases, the other tends to decrease.
Features with high positive or negative correlations might be related and could potentially impact each other.
Identifying strong correlations is crucial for feature selection and understanding which features might be redundant or highly influential in predicting churn.
It's important to note that correlation does not imply causation. Further analysis and domain knowledge are necessary to draw meaningful conclusions about the relationships between variables and their impact on the target variable (churn in this case).

```python
# Visualization 4: Customer Service Calls Distribution for Churned Users
plt.figure(figsize=(10, 6))
sns.countplot(x='Customer service calls', data=df[df['Churn'] == 1], palette='viridis')
plt.title('Customer Service Calls Distribution for Churned Users')
plt.xlabel('Number of Customer Service Calls')
plt.ylabel('Count')
plt.show()
```

Conclusion:

The higher concentration of churned users with multiple customer service calls suggests a potential correlation between dissatisfaction with the service (indicated by the need for multiple service calls) and the decision to churn.
Further analysis and possibly customer feedback data could provide insights into the reasons behind the increased number of customer service calls for churned users.
This information could be valuable for the telecom company to improve customer service and reduce churn.
